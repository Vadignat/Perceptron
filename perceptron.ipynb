```python
import loader as l
from matplotlib import pyplot as plt
```

```python
ld = l.loader()
tri = ld.getTrainInp()
tro = ld.getTrainOut()
tsi = ld.getTestInp()
tso = ld.getTestOut()
```

```python
from MLPerceptron import MLP

p = MLP(tri, tro, tsi, tso, neuronNum=(12,30))
err_tr, err_ts = p.learn(eta = 0.025,epoches=3000)
```

```python
f1 = plt.figure(1)
fa1 = f1.add_subplot(1, 1, 1)

out = p.calc(tri)
fa1.plot(tri, out, "bo")
fa1.plot(tri, tro, "r+")
out = p.calc(tsi)
fa1.plot(tsi, out, "gv")
fa1.plot(tsi, tso, "y+")
f2 = plt.figure(2)
fa2 = f2.add_subplot(1,1,1)
err_x = [i for i in range(1, len(err_tr) + 1)]
fa2.plot(err_x,err_tr, "b-")
fa2.plot(err_x,err_ts, "r-")
plt.show()
```
